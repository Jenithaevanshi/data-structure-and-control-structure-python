### Exercise
Write a function that takes a list of prices of items purchased and calculates the total cost. Assume the tax rate as `tax_rate = 0.1` and discount as `discount = 0`. The total cost is given by `total cost = subtotal * (1 + tax_rate) - discount`.

````python
def calc_tot_cost(items, tax_rate = 0.1, discount = 0):
    subtotal = sum(items)
    total = subtotal * (1 + tax_rate) - discount
    return total
````

````python
round(calc_tot_cost([100.50, 254.22, 39.95, 15.50, 49.99, 29.99, 42.75, 19.99, 59.95, 34.50]), 2)
````

### Exercise
Create a function that takes two tuple containing the `x` and `y` coordinate values of two points and computes the Euclidean distance

````python
from math import sqrt
````

````python
def calc_dist(point_1, point_2):
    x1, y1 = point_1
    x2, y2 = point_2

    return sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
````

````python
calc_dist((1, 2), (4, 6))
````
